## import library

```python
import kagglehub
import os
import pandas as pd
```

## inisiasi dataset dari kaggle

```python
path = kagglehub.dataset_download("prepinstaprime/europe-bike-store-sales")

data = pd.read_csv(os.path.join(path, 'Sales.csv'))
df = pd.DataFrame(data.values)
df
```

## setting data ke dalam template

```python
d = df[0:][[0, 1, 2, 3, 4]]
d.columns = ["Date", "Day", "Month", "Year", "Customer_Age"]
d
```

## descending kolom Date
