## import library yang digunakan

```python
import pandas as pd
import kagglehub
import os
```

## baca dataset

```python
path = kagglehub.dataset_download("prepinstaprime/europe-bike-store-sales")
data = pd.read_csv(os.path.join(path, 'Sales.csv'))
df = pd.DataFrame(data.values)
df.columns = data.columns
df
```

## transformasi data Month

```python
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df['Month'] = df['Month'].map(month_map)
df
```

## Kategori beberapa kolom

```python
# Kategori umur lebih detail
df['Age_Category'] = pd.cut(df['Customer_Age'].astype(int), bins=[0, 18, 25, 35, 50, 65, 100], 
                            labels=['Child', 'Youth', 'Young Adult', 'Adult', 'Senior', 'Elderly'])

# Kategori profit (untung/rugi)
df['Profit_Category'] = df['Profit'].astype(int).apply(lambda x: 'Profit' if x > 0 else 'Loss')

# Kategori revenue (rendah, sedang, tinggi)
df['Revenue_Category'] = pd.qcut(df['Revenue'].astype(int), q=3, labels=['Low', 'Medium', 'High'])

# Kategori berdasarkan jenis produk utama
df['Main_Product_Category'] = df['Product_Category'].apply(lambda x: x.split()[0] if isinstance(x, str) else x)

df
```
