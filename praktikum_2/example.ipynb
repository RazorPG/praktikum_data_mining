## import library yang digunakan

```python
import pandas as pd
from openpyxl import load_workbook
```

## inisiasi file excel

```python
wb = load_workbook('D:\\Studied_at_UNP\\SEMESTER-5\\praktikum_data_mining\\praktikum_2\\Data Alumni.xlsx')
sheet_range = wb['Sheet1']
df = pd.DataFrame(sheet_range.values)
df
```

## seleksi data sampling

```python
df.columns = df.iloc[0]
df = df.drop(0)
df.loc[:, ["NIM", "TTL"]]
```

## seleksi data untuk mencari umur

```python
# membuat kolom baru untuk tahun lahir
for index, row in df.iterrows():
    df.loc[index, 'Tahun Lahir'] = row['TTL'][-4:]

# import datetime untuk mendapatkan tahun sekarang
from datetime import datetime
now = datetime.now()

now.year
for index, row in df.iterrows():
    df.loc[index, "Umur"] = now.year - int(row["Tahun Lahir"])
df

```

## seleksi data untuk mencari masa studi

```python
for index, row in df.iterrows():
    df.loc[index, "Masa Studi"] =row["Tahun Lulus"] - row["Tahun Masuk"]
df
```
