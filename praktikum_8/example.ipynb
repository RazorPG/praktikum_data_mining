## load library yang akan digunakan

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

## membaca data kelulusan mahasiswa

```python
mhs = pd.read_excel(r'D:\Studied_at_UNP\SEMESTER-5\praktikum_data_mining\praktikum_8\Data Kelulusan Mahasiswa.xlsx')
mhs.head(10)
```

## Menyiapkan Variabel Fitur (X)

```python
x = mhs.drop(["StatusKelulusan"], axis=1)
x.head(5)
```

## Menyiapkan variabel target (y = StatusKelulusan)

```python
y = mhs["StatusKelulusan"]
y.head(10)
```

## memanggil fungsi Gaussian Naive Bayes

```python
nbc = GaussianNB()
```

## Memasukkan data training pada fungsi klasifikasi naive bayes

```python
data_training = nbc.fit(x, y)
```

## melakukan prediksi pada data training

```python
y_predict = data_training.predict(x)
print(y_predict)
```

## Menyiapkan Data Testing

```python
Jenis_Kelamin = 1
Status_Kerja = 0
Umur = 25
Status_Nikah = 1
IPS1 = 3.12
IPS2 = 2.75
IPS3 = 2.05
IPS4 = 3.01
IPS5 = 2.56
IPS6 = 3.67
IPS7 = 2.75
IPS8 = 3.45
IPK = np.average([IPS1, IPS2, IPS3, IPS4, IPS5, IPS6, IPS7, IPS8])
Data_Testing = [[Jenis_Kelamin, Status_Kerja, Umur, Status_Nikah, IPS1, IPS2, IPS3, IPS4, IPS5, IPS6, IPS7, IPS8, IPK]]
```

## mencetak hasil prediksi berdasarkan kelulusan mahasiswa

```python
# lakukan prediksi untuk Data_Testing yang sudah disiapkan
pred = data_training.predict(Data_Testing)
pred_label = int(pred[0]) if hasattr(pred, "__len__") else int(pred)

if pred_label == 0:
    hasil = "Terlambat"
elif pred_label == 1:
    hasil = "Tepat"
else:
    hasil = "Error"

print("Hasil Prediksi kelulusan mahasiswa adalah :", hasil)
```

## Menghitung dan Menampilkan Nilai Akurasi

```python
print("Nilai Akurasi = %0.2f" % accuracy_score(y, y_predict))
```

## laporan klasifikasi

```python
print(classification_report(y, y_predict))
```
